```python
from pathlib import Path
import os
```

```python
path = Path('/home/ehsan/CVAssignment/Footages/Frames')
def get_files(path):
    flist = [p for p in Path(path).iterdir() if p.is_file()]
    return flist
```

```python
get_files(path)
```

```python
path = '/home/ehsan/CVAssignment/Footages/Frames'
f_paths = []
for root,d_names,f_names in os.walk(path):
    for f_name in f_names:
        if f_name.endswith('.jpg'):
            f_path = os.path.join(path, f_name)
            f_paths.append(f_path)
```
