```python
import cv2
from IPython.core.display import display, HTML
import matplotlib.pyplot as plt
from tqdm.notebook import trange, tqdm
import numpy as np
import os
import matplotlib.animation as animation
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# display a video
def get_video(frames, interval=50, figsize=(6,4)): #create a video to play in notebook
    
    def animate(i):
        img.set_data(frames[i])
        return (img,)

    fig, ax = plt.subplots(1,1,figsize=figsize)
    img = ax.imshow(frames[0])
    ax.axis('off')
    anim = animation.FuncAnimation(fig, animate, frames=len(frames), interval=interval, blit=True, repeat_delay=5000)
    plt.close()
    return anim.to_html5_video()
```

```python
# converting extracted face_frames to videos
def draw_bbox(path, labels,frames=None):
    if frames is None:
        frames = []

    for dirName, subdirList, fileList in os.walk(path):
        for fname in fileList:
            frame = cv2.imread(os.path.join(path, fname)) # read a frame
            frame = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB) # BGR2RGB
            fname = float(fname[:-3]) # find a file_name and convert it to digits
            label = labels.loc[labels["im_name"]==fname] # finding the label of the image
            x = labels.iloc[0,2]
            y = labels.iloc[0,3]
            w = labels.iloc[0,4]
            h = labels.iloc[0,5]
            frame = cv2.rectangle(frame, (int(x-w/2), int(y+h/2)), (int(x+w/2), int(y-w/2)), (0, 0, 255), 1)
            frames.append(frame)
    return frames
```

```python
labels = pd.read_csv('./RainyDay2_tracked_132.csv',header=None, delim_whitespace=True,
                     names=['im_name', 'class_id', 'seq', 'Xc','Yc','w','h','?'])
labels = labels.drop(['seq', '?'], axis=1)
labels.head()
```

```python
path = './SampleData/'
frames = draw_bbox(path=path,labels=labels)
```

```python
HTML(get_video(frames,interval=400))
```
