```python
def show_img(im, ax=None, figsize=(8,8), title=None):
    if not ax: _,ax = plt.subplots(1,1,figsize=figsize)
    if len(im.shape)==2: im = np.tile(im[:,:,None], 3) 
    ax.imshow(im);
    ax.xaxis.set_visible(False)
    ax.yaxis.set_visible(False)
    if title: ax.set_title(title)
    return ax
```

```python
def draw_rect(ax, xy, w, h):
    patch = ax.add_patch(patches.Rectangle(xy, w, h, fill=False, edgecolor='yellow', lw=2))
    patch.set_path_effects([patheffects.Stroke(linewidth=3, foreground='black'), patheffects.Normal()])
```

```python
class MYdataset(Dataset):
    def __init__(self,csv_file, root_dir,c2i, tfms=None):
        super().__init__()
        self.frames_list = pd.read_csv(csv_file)
        self.root_dir = root_dir
        self.c2i = c2i
        self.tfms = tfms
        
        
    def __len__(self):
        return len(self.frames_list)
    

    def __getitem__(self, idx):
          
        img_path = os.path.join(root_dir,frames_list.iloc[idx,0])
        img = PIL.Image.open(img_path)
       
        Y = frames_list.iloc[idx,1]
        xc = frames_list.iloc[idx,2]
        yc = frames_list.iloc[idx,3]
        w = frames_list.iloc[idx,4]
        h = frames_list.iloc[idx,5]
        
        if tfms is not None:
            img = self.tfms(img)
            
        return img,(Y,xc,yc,w,h)
```
