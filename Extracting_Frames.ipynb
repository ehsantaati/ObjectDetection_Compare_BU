```python
cap = cv2.VideoCapture('Footages/RainyDay1.mp4')
_,image = cap.read()
#length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
length =500
count = 0
for i in trange(length):
    cv2.imwrite("frame%d.jpg" % count, image)
    _,image = cap.read()
    count += 1
```
